```python
#. What is the population of the top 5 most populous countries
top_5_countries = df.sort_values(by='2022 Population', ascending=False).head(5)

print(top_5_countries[['Country/Territory', '2022 Population']])
```

```python
##. Create a new column showing GDP per capita ( GDP / Population )

# Example: Adding a dummy GDP column for practice (First 5 rows ke liye hi fill kar raha hoon)
dummy_gdp = [17000000000000, 3200000000000, 23000000000000, 1200000000000, 400000000000] + [None]*(len(df)-5)

df['GDP'] = dummy_gdp

# Calculate GDP per Capita (where GDP is not null)
df['GDP per Capita'] = df['GDP'] / df['2022 Population']

print(df[['Country/Territory', 'GDP', '2022 Population', 'GDP per Capita']].head(10))
```

```python
#. Filter out countries with population > 100 million and area < 500,000 sq km.
filtered_countries = df[(df['2022 Population'] > 100000000) & (df['Area (km²)'] < 500000)]
print(filtered_countries)

```

```python
print(filtered_countries[['Country/Territory', '2022 Population', 'Area (km²)']])
```

```python
#. Sort the countries based on density in descending order
df['density'] = df['2022 Population'] / df['Area (km²)']

sorted_countries=df.sort_values(by='density',ascending=False)
print(sorted_countries)
```
