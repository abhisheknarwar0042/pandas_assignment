```python
#convert the date_added column to datetime and extract the year and month

# Load dataset
df = pd.read_csv('netflix_titles.csv.zip')

# Safely convert date column (handle NaN)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year and month
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Show result
print(df[['date_added', 'year_added', 'month_added']])

```

```python
#7. How many movies were released in 2019?


movies_2019 = df[(df['release_year'] == 2019) & (df['type'] == 'Movie')]
print(movies_2019)
```

```python
#. Group by rating and show the count of each category

rating_counts = df.groupby('rating').size().reset_index(name='count')


print(rating_counts)


```
