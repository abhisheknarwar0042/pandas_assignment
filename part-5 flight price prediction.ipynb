```python
# Most frequent airline
most_frequent_airline = df['airline'].value_counts().idxmax()
print("Most frequent airline:", most_frequent_airline)

# Most frequent source city
most_frequent_source = df['source_city'].value_counts().idxmax()
print("Most frequent source city:", most_frequent_source)
```

```python
# Check duplicates
print("Total duplicates before removing:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

# Check for missing values
print("Missing values per column:\n", df.isnull().sum())

# Fill missing values (simple method: forward fill or mean, depends on column type)
# Example:
df = df.fillna(method='ffill')  # Forward fill (you can use other methods also)

# After cleaning
print("Missing values after handling:\n", df.isnull().sum())
```
